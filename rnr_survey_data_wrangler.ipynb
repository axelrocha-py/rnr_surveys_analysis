On this notebook I analyze the customer satisfaction surveys from Rock n Ribs restaurant since October 2019 until May 2020 looking up trends or variations from the 1st semester operation of the year. The dataset is given by Survey Monkey platform and the aim of this analysis is cross this data with Facebook campaigns dataset to find hidden patterns. 

```python
import pandas as pd
import numpy as np
```

<b>Note: </b>From data privacy issues, I only show the firstly three rows of my dataset anytime. But you can look that this dataset has 4845 useful rows.

```python
df_survey = pd.read_csv('./survey_05_05_2020.csv')
df_survey[0:4]
```

```python
df_survey.columns
```

```python
# Cleaning column's names
df_survey.rename(columns={'¿En qué Franquicia ROCK n RIBS te encuentras?':'Franquicia',
        ' ¿Qué tan probable es que recomiendes ROCK n RIBS a tus amigos o colegas?':'Recomendacion',
        '¿Qué te pareció la comida en general? Seleccione 1 respuesta.': 'Comida',
        '¿Qué te parecieron las bebidas en general? Seleccione 1 respuesta': 'Bebida',
        '¿Cómo calificaría la relación calidad precio de nuestros alimentos y bebidas.?':'CalidadPrecio',
        '¿Qué tan atento ha sido su mesero ante sus necesidades durante su visita en ROCK n RIBS?':'AtencionMesero(E)',
        'Unnamed: 15':'AtencionMesero(B)',
        'Unnamed: 16':'AtencionMesero(R)',
        'Unnamed: 17':'AtencionMesero(M)',
        'Unnamed: 18':'NombreMeseroVacio',
        '¿Cúal es el nombre de su mesero?':'NombreMesero', 
        '¿El (La) Gerente se acercó en algún momento a su mesa y estuvo atento(a) a sus necesidades?':'AtencionGerente',
        '¿Cuál el nombre de su Gerente?':'NombreGerente',
        '¿Cuánto tiempo ha sido cliente de ROCK n RIBS?':'TiempoSerCliente',
        '¿Cómo te enteraste de nosotros?':'MedCon',
        'Unnamed: 24':'OtroMedCon',
        '¿Tienes algún otro comentario, pregunta, queja, sugerencia o inquietud?':'Comentarios',
        'Información de contacto':'NombreCliente',
        'Unnamed: 27':'OcupacionVacio',
        'Unnamed: 28':'EmailVacio',        
        'Unnamed: 29':'Cumpleanos',                          
        'Unnamed: 30':'CiudadCliente',
        'Unnamed: 31':'CiudadDeseoFranqVacio',                          
        'Unnamed: 32':'CP',  
        'Unnamed: 33':'Pais',                            
        'Unnamed: 34':'Email',                         
        'Unnamed: 35':'Celular',
        '¿TE INTERESA UNA FRANQUICIA ROCK n RIBS?':'DeseoFranq(SI)',
        'Unnamed: 37':'DeseoFranq(NO)',
        'Unnamed: 38':'CiudadDeseoFranq'}, inplace = True)
df_survey.head()
```

```python
df_survey.drop([0],axis=0, inplace=True)
df_survey[0:3]
```

```python
df_survey.shape
```

```python
# Verifying null values
df_survey.isna().sum()
```

```python
# Cleaning our dataset of null mayority columns
df_survey.drop(['email_address', 
                'first_name', 
                'last_name', 
                'custom_1', 
                'OcupacionVacio', 
                'EmailVacio', 
                'CiudadCliente',
                'CiudadDeseoFranqVacio',
                'CP',
                'Pais'], axis=1, inplace=True)
df_survey.isna().sum()
```

```python
# Unifying AtencionMesero divided columns
am_e = np.array(df_survey['AtencionMesero(E)'])
am_b = np.array(df_survey['AtencionMesero(B)'])
am_r = np.array(df_survey['AtencionMesero(R)'])
am_m = np.array(df_survey['AtencionMesero(M)'])

am_list = []
for i in range(0, (df_survey.shape[0])):
    if am_e[i] == 'Excelente':
        am_list.append('Excelente')
    else:
        if am_b[i] == 'Bueno':
            am_list.append('Bueno')
        else:
            if am_r[i] == 'Regular':
                am_list.append('Regular')
            else:
                if am_r[i] == 'Malo':
                    am_list.append('Malo')
                else:
                    am_list.append(None)
len(am_list)
```

```python
df_survey['AtencionMesero'] = am_list
```

```python
df_survey.drop(['AtencionMesero(E)', 'AtencionMesero(B)', 'AtencionMesero(R)', 'AtencionMesero(M)'],axis=1, inplace=True)
```

```python
df_survey.columns
```

```python
df_survey.count()
```

```python
# Unifying DeseoFranq divided columns
df_s = np.array(df_survey['DeseoFranq(SI)'])
df_n = np.array(df_survey['DeseoFranq(NO)'])
```

```python
df_list = []
for i in range(0, (df_survey.shape[0])):
    if df_s[i] == 'SÍ':
        df_list.append('SI')
    else:
        df_list.append('NO')

len(df_list)
```

```python
df_survey['DeseoFranq'] = df_list
df_survey.drop(['DeseoFranq(SI)', 'DeseoFranq(NO)'],axis=1, inplace=True)
```

```python
df_survey.columns
```

```python
df_survey.count()
```

```python
# Unifying DeseoFranq divided columns
nm_c = np.array(df_survey['NombreMesero'])
nm_v = np.array(df_survey['NombreMeseroVacio'])

nm_list = []
for i in range(0, (df_survey.shape[0])):
    if type(nm_c[i])==str:
        nm_list.append(nm_c[i])
    else:
        if type(nm_v[i]==str):
            nm_list.append(nm_v[i])
        else:
            nm_list.append(None)

len(nm_list)
```

```python
df_survey['NomMesero'] = nm_list
df_survey.drop(['NombreMesero', 'NombreMeseroVacio'],axis=1, inplace=True)
```

```python
df_survey[0:3]
```

```python
# Organize the columns
df_survey = df_survey[['respondent_id', 
                       'collector_id',
                       'date_created',
                       'date_modified',
                       'ip_address',
                       'NombreCliente',
                       'Email',
                       'Celular',
                       'Franquicia',
                       'Recomendacion',
                       'Comida',
                       'Bebida',
                       'CalidadPrecio',
                       'TiempoSerCliente',
                       'MedCon',
                       'OtroMedCon',
                       'AtencionMesero',
                       'NomMesero',
                       'AtencionGerente',
                       'NombreGerente',
                       'DeseoFranq',
                       'CiudadDeseoFranq',
                       'Comentarios']]
df_survey[0:3]
```
